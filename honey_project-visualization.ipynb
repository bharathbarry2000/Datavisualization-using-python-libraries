# 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings

```python
P=sns.distplot(h_df.priceperlb,bins=20,kde=False,color='g')
plt.ylabel('Count of states')
plt.xlabel('Price per Pound.')
plt.show()
```

# 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

# 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='year',y='prodvalue',data=h_df)
plt.show()
```

* We can observe there is many outlier's.
* The median Production value for all the years lies between 0 to 0.5
* In the recent years the productionvalue has rapidly increased.
* The year 2010 have highest maximum productionvalue 

# 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')


```python
plt.figure(figsize=(10,10))
sns.pairplot(h_df[['numcol', 'yieldpercol', 'totalprod', 'prodvalue','year']])
plt.show()
```

* Total production - Number of colonies 
    * We can observe that there is a positive linear relationship between "total production" & "number of colonies".
    * If a state have more number of colonies then the total production of that state is increasing.
    
* production value - Number of colonies 
    * We can observe that there is a positive linear relationship between "production value" & "number of colonies".
    * If a state have more number of colonies then the total production value of that state is increasing.
    
* production value - Total production 
    * We can observe that there is a positive linear relationship between "total production" & "Production value".
    * If a state have more production then the production value is increasing.
